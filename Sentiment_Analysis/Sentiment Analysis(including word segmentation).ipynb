```python
############# 蔡英文

import re
import pandas as pd
import jieba as jb
import jieba.analyse

dictionary = pd.read_excel("台大情感分析詞典.xlsx")
dictionary_to_list = dictionary.values.tolist()

Tsai_nickname = pd.read_excel('(民)蔡英文綽號.xlsx')
nickname_to_list = Tsai_nickname.values.tolist()
for namelist in nickname_to_list:
    #print(name)
    #print(type(name))
    for name in namelist:
        #print('測試:', test)
        #print(type(test))
        jb.suggest_freq('%s'%(name), True)

adv = pd.read_excel('程度副詞.xlsx')
adv_tolist = adv.values.tolist()
#print(adv_tolist)
adv_list = []
for adv_word in adv_tolist:
    adv_dictionary = [adv_word[1], adv_word[2]]
    adv_list.append(adv_dictionary)

Tsai = pd.read_excel('蔡英文八卦版(大選一個月內)_DataSet_final.xlsx', encoding = 'utf-8')
Tsai_valuetolist = Tsai.values.tolist()
phrase_list = []

#print('-'*10)
#print(adv_list)
#print('-'*10)
#print(dictionary_to_list)
#print('-'*10)
        
def word_score(wordlist):
    total_score = 0
    sentiment_score = 0
    adv_quality = 1
    
    for word in wordlist:
        #print('斷詞:', word)
        
        for adv_word, adv_score in adv_list:
            if word == adv_word:
                adv_quality = adv_score
            
        
        for sentiment_word in dictionary_to_list:
            if word == sentiment_word[0]:                    
                #print(sentiment_word)                        
                #print('抓到的字:%s %s'%(word, sentiment_word[1]))
                sentiment_score += sentiment_word[1]
    
    total_score += sentiment_score*adv_quality
    #print(total_score, sentiment_score, adv_quality)    
    return total_score
        
    
for article in Tsai_valuetolist:
    #print(article[1]) #文章
    for nickname in nickname_to_list:
        #print(nickname) #綽號
        try:
            phrases = re.findall(r'[\u4e00-\u9fa5%0-9]*{}[\u4e00-\u9fa5%0-9]*'.format(nickname), article[1]) ##正則表達式
            #print('==============', phrases)
            if phrases:
                #print(phrases)
                author_phrase = [article[0], phrases]
                phrase_list.append(author_phrase)
            break
        except:
            continue
            #print(article)
#print(phrase_list)

for Tsai_phrase in phrase_list:
    #print(Tsai_phrase)
    #print('-'*10)
    phrase_total_score = 0
    
    for single_phrase in Tsai_phrase[1]:
        #print(single_phrase)
        words = jb.cut(single_phrase)
        #print("/".join(words)) #jeiba斷詞(有毒--需要註解掉)
        phrase_total_score += word_score(words)
                   
    print('集合:%s，分數:%s'%(Tsai_phrase, phrase_total_score))
    print('='*5)
```
